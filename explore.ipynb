```python
all_sql_tables = pd.read_sql("""SELECT * FROM sqlite_master""", con2)
table_names = all_sql_tables[all_sql_tables['type'] == 'table']
table_names = table_names['tbl_name'].values
table_names
```

```python
data_frames = []
for name in table_names:
    df = pd.read_sql(f"""SELECT * FROM {name}""", con2)
    data_frames.append(df)
```

```python
for name, df in zip(table_names, data_frames):
    csv_name = f'data/{name}.csv'
    df.to_csv(csv_name)
    # write code for r
    print(f"{name} <- read_csv(\"{csv_name}\")")
```
